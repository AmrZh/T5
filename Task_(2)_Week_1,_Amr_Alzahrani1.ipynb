## 1. Exploratory Data Analysis (EDA)

### Perform summary statistics on the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Task (2) Dataset.csv')
df.head()
```

```python
df.describe()
```

### Identify and analyze patterns in the data.

```python
reviews = df['TravelTimeHistoric'].value_counts ()
reviews
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
sns.histplot(data=df, x='TrafficIndexLive')
plt.title('Distribution of Traffic Index')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='TrafficIndexWeekAgo', y='TrafficIndexLive', data=df)
plt.title('Traffic Index Week Ago vs. Traffic Index Live')
plt.xlabel('Traffic Index Week Ago')
plt.ylabel('Traffic Index Live')
plt.show()
```

## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
sns.countplot(data=df, x='TrafficIndexLive', palette='viridis')
plt.xticks(rotation=45)
plt.show()
```
